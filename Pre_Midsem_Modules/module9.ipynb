```python
from dataclasses import dataclass, field
from datetime import datetime
from typing import Optional

# Output Model

@dataclass(frozen=True)
class SavingsEstimationRecord:
    """Quantified savings."""
    kw_saved: float                  # kW
    energy_saved_kwh: float          # kWh
    rupees_saved_per_interval: float # ₹
    timestamp: datetime = field(default_factory=datetime.now)

# Module Implementation

class EnerviaSavingsEngine:
    """Savings Engine."""

    def __init__(self):
        # State tracking
        self.cumulative_kw_saved = 0.0
        self.cumulative_rupees_saved = 0.0

    def compute_savings(
        self, 
        baseline_power: float, 
        reduced_power: float, 
        duration_hours: float, 
        tariff: float
    ) -> SavingsEstimationRecord:
        """Compute savings."""
        
        # Calculate Power Delta
        kw_saved = max(0.0, baseline_power - reduced_power)
        
        # Calculate Energy Delta
        energy_saved_kwh = kw_saved * duration_hours
        
        # Calculate Financial Delta
        rupees_saved = energy_saved_kwh * tariff
        
        # Update Internal State
        self.cumulative_kw_saved += kw_saved
        self.cumulative_rupees_saved += rupees_saved

        return SavingsEstimationRecord(
            kw_saved=round(kw_saved, 2),
            energy_saved_kwh=round(energy_saved_kwh, 2),
            rupees_saved_per_interval=round(rupees_saved, 2)
        )

# savings_engine = EnerviaSavingsEngine()
# report = savings_engine.compute_savings(
#     baseline_power=500.0,
#     reduced_power=450.0,
#     duration_hours=2.0,
#     tariff=8.5
# )
```
